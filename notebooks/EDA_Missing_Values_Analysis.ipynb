# EDA: Missing Values Analysis

This notebook performs a comprehensive missing value analysis on the interim kidney disease dataset.

## 1. Load and Explore the Dataset

```python
import pandas as pd
import numpy as np

# Load interim dataset
df = pd.read_csv("../data/interim/kidney_disease_interim.csv")

print("Dataset loaded successfully!")
print(f"\nDataset shape: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"\nFirst few rows:")
print(df.head())
```

```python
# Display basic dataset information
print("Dataset Information:")
print("="*70)
print(f"Total rows: {df.shape[0]}")
print(f"Total columns: {df.shape[1]}")
print(f"\nColumn names: {df.columns.tolist()}")
print(f"\nData types:")
print(df.dtypes)
```

## 2. Calculate Missing Values

```python
# Calculate missing values per column
missing_counts = df.isnull().sum()
missing_percentages = (df.isnull().sum() / len(df)) * 100
total_cells = np.prod(df.shape)
total_missing = missing_counts.sum()

print("Missing Values Summary:")
print("="*70)
print(f"Total cells in dataset: {total_cells:,}")
print(f"Total missing values: {total_missing:,}")
print(f"Percentage of total missing: {(total_missing / total_cells) * 100:.2f}%")
print(f"\nColumns with missing data: {(missing_counts > 0).sum()}")
print(f"Columns without missing data: {(missing_counts == 0).sum()}")
```

## 3. Create Missing Values Table

```python
# Create comprehensive missing values table
missing_df = pd.DataFrame({
    'Column': df.columns,
    'Missing Count': missing_counts.values,
    'Missing Percentage': missing_percentages.values,
    'Data Count': len(df) - missing_counts.values,
    'Data Percentage': 100 - missing_percentages.values
})

# Sort by missing percentage (descending)
missing_df_sorted = missing_df.sort_values('Missing Percentage', ascending=False)

print("\nMissing Values Table (sorted by missing percentage):")
print("="*90)
print(missing_df_sorted.to_string(index=False))
print("="*90)
```

```python
# Summary and insights based on the computed missing-values analysis

# Core metrics
n_rows, n_cols = df.shape
dataset_missing_pct = (total_missing / total_cells) * 100
cols_with_missing = int((missing_counts > 0).sum())
cols_without_missing = int((missing_counts == 0).sum())

# Severity buckets
considerable_missing = missing_df_sorted[missing_df_sorted["Missing Percentage"] > 10]
acceptable_missing = missing_df_sorted[
    (missing_df_sorted["Missing Percentage"] > 0) &
    (missing_df_sorted["Missing Percentage"] <= 10)
]

top5 = missing_df_sorted.head(5)

print("SUMMARY")
print("=" * 80)
print(f"Dataset size: {n_rows} rows × {n_cols} columns")
print(f"Total missing values: {int(total_missing):,} ({dataset_missing_pct:.2f}% of all cells)")
print(f"Columns with missing data: {cols_with_missing}/{n_cols}")
print(f"Columns without missing data: {cols_without_missing}/{n_cols}")

print("\nTop 5 columns by missing percentage:")
print(top5[["Column", "Missing Count", "Missing Percentage"]].to_string(index=False))

print("\nINSIGHTS")
print("=" * 80)
if dataset_missing_pct == 0:
    print("- No missing values detected; dataset is complete.")
elif dataset_missing_pct <= 10:
    print("- Overall missingness is acceptable and likely manageable with simple imputation.")
else:
    print("- Overall missingness is considerable; careful feature selection and robust imputation are required.")

print(f"- Considerable-missing columns (>10%): {len(considerable_missing)}")
print(f"- Acceptable-missing columns (>0% to <=10%): {len(acceptable_missing)}")

if len(considerable_missing) > 0:
    print("\n- Columns with considerable missingness should be reviewed for potential exclusion or imputation:")
    print("  " + ", ".join(considerable_missing["Column"].tolist()))
```
