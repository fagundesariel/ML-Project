# Normalization
This notebook focuses on normalizing numeric features in the dataset.
Normalization is a crucial step in data preprocessing, especially for algorithms that are sensitive to the scale of features (e.g., KNN, SVM, neural networks).
We will apply Min-Max normalization to scale numeric features to a range of [0, 1].

We will load the kidney disease dataset, identify numeric features, and apply Min-Max normalization to these features.

## 1. Load and Prepare Data
We will load the interim dataset and identify numeric features for normalization.

```python
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Load interim dataset
interim_path = Path("../data/interim/kidney_disease_interim.csv")

df = pd.read_csv(interim_path)
```

# Identify numeric columns

```python
numeric_cols = df.select_dtypes(include='number')
print("Numeric columns identified for normalization:")
for col in numeric_cols:
    print(f"  - {col}")
```

## 2. Apply Min-Max Normalization
We will apply Min-Max normalization to the identified numeric features, scaling them to a range of [0, 1].

```python
# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the numeric columns
df_normalized = df.copy()
df_normalized[numeric_cols.columns] = scaler.fit_transform(numeric_cols)

# Display the range of values for each numeric column after normalization
print("\nRange of values for numeric columns after Min-Max normalization:")
for col in numeric_cols.columns:
    col_min = df_normalized[col].min()
    col_max = df_normalized[col].max()
    print(f"  - {col}: min={col_min:.2f}, max={col_max:.2f}")
```

# Display the first few rows of the normalized dataset

```python
print("\nFirst few rows of the normalized dataset:")
print(df_normalized.head())
```

# Export the normalized dataset to a new CSV file

```python
# Resolve project root whether kernel starts in repo root or notebooks/
project_root = Path.cwd()
if not (project_root / "data").exists() and (project_root.parent / "data").exists():
    project_root = project_root.parent

processed_dir = project_root / "data" / "processed"
processed_dir.mkdir(parents=True, exist_ok=True)

normalized_path = processed_dir / "kidney_disease_normalized.csv"
df_normalized.to_csv(normalized_path, index=False)
print(f"Normalized dataset saved to: {normalized_path.resolve()}")
```
